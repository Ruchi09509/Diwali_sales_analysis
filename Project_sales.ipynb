```python
# import python libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df= pd.read_csv('C:/Python_Diwali_Sales_Analysis-main/Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
# display information about the dataset
df.info()
```

```python
# display the shape of dataset
print("Number of columns and row in the dataset :",df.shape)
```

```python
# display data using head method 
# by default it show 5 data entries
df.head(7)
```

```python
# Slice the 9 rows from 10th
df[8:18]
```

```python
# drop unrelated/blank columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
# find the null values of dataset
df.isna().sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
# display the names of columns in the dataset
df.columns
```

```python
#rename column
df.rename(columns= {'Marital_Status':'Shaadi'})
```

```python
# Exploratory Data Analysis
```

```python
# Gender
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df, color='red')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

```python
#Result
From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#Age
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age, color='purple')
```

```python
#Result
From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#Sales
```

```python
# Group by 'State', sum orders, sort in descending order, and select top 10 states
sales_state = df.groupby('State', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (15, 5)})

# Create histogram-like bar plot
ax = sns.histplot(data=sales_state, x='State', weights='Orders', bins=len(sales_state), 
                   discrete=True, stat="count", color='royalblue')

# Annotate each bar with the order value
for p in ax.patches:
    ax.annotate(f"{int(p.get_height())}", 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold')

# Add labels and title
plt.xlabel("State", fontsize=14)
plt.ylabel("Total Orders", fontsize=14)
plt.title("Histogram of Total Orders from Top 10 States", fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=12)

# Show plot
plt.show()
```

```python
#Result
From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
#Marital Status
```

```python
# Count occurrences of each marital status category
marital_counts = df['Marital_Status'].value_counts().reset_index()
marital_counts.columns = ['Marital_Status', 'Count']

# Set figure size
sns.set(rc={'figure.figsize': (7, 5)})

# Create scatter plot
ax = sns.scatterplot(data=marital_counts, x='Marital_Status', y='Count', size='Marital_Status', hue='Marital_Status', palette='viridis', sizes=(50, 300))

# Annotate points with values
for i in range(marital_counts.shape[0]):
    plt.text(x=marital_counts['Marital_Status'][i], y=marital_counts['Count'][i] + 1, 
             s=marital_counts['Count'][i], ha='center', fontsize=12)

# Add labels and title
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.title("Scatter Plot of Marital Status Distribution")

# Show plot
plt.show()

```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

```python
#Result
From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
#Occupation
```

```python
# Count occurrences of each occupation
occupation_counts = df['Occupation'].value_counts()

# Create figure and axis
plt.figure(figsize=(12, 6))

# Generate line plot
plt.plot(occupation_counts.index, occupation_counts.values, marker='o', linestyle='-', 
         color='b', linewidth=2, markersize=8, label="Occupation Count")

# Add labels and title
plt.xlabel("Occupation", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Occupation Distribution (Line Plot)", fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
# Group by 'Occupation' and sum the 'Amount', then sort in descending order
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size
plt.figure(figsize=(20, 5))

# Create area plot
plt.fill_between(sales_state['Occupation'], sales_state['Amount'], color='skyblue', alpha=0.6)

# Add a line plot for better readability
plt.plot(sales_state['Occupation'], sales_state['Amount'], marker='o', linestyle='-', 
         color='b', linewidth=2, markersize=8, label="Total Sales Amount")

# Add labels and title
plt.xlabel("Occupation", fontsize=14)
plt.ylabel("Total Sales Amount", fontsize=14)
plt.title("Total Sales Amount by Occupation (Area Chart)", fontsize=16, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add grid for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show legend
plt.legend()

# Show plot
plt.show()
```

```python
#Result
From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
#Product Categories
```

```python
# Count occurrences of each product category
category_counts = df['Product_Category'].value_counts().reset_index()
category_counts.columns = ['Product_Category', 'Count']

# Set figure size
plt.figure(figsize=(12, 8))

# Create horizontal bar chart with `hue` assigned to `y`
ax = sns.barplot(y="Product_Category", x="Count", data=category_counts, hue="Product_Category", 
                 palette="viridis", legend=False)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Add labels and title
plt.xlabel("Count", fontsize=14)
plt.ylabel("Product Category", fontsize=14)
plt.title("Product Category Distribution (Horizontal Bar Chart)", fontsize=16, fontweight='bold')

# Show plot
plt.show()

```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
#Result
From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# Group by 'Product_ID', sum 'Orders', sort in descending order, and select top 10 products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Define labels and values
labels = sales_state['Product_ID']
sizes = sales_state['Orders']

# Set figure size
plt.figure(figsize=(10, 6))

# Create pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, 
        colors=plt.cm.viridis(np.linspace(0, 1, len(labels))), wedgeprops={'edgecolor': 'black'})

# Add title
plt.title("Top 10 Products by Total Orders (Pie Chart)", fontsize=14, fontweight='bold')

# Show plot
plt.show()
```

```python
#over all output of this analysis
Married women age group 26-35 years from UP, Maharastra and Karnataka working in IT, Healthcare and 
Aviation are more likely to buy products from food ,clothing and Electronics category.
```
